```python
data = pd.read_csv('combined.csv')
```

# Process data: extract rating, 

```python
rating = data['rating']
```

```python
genres = data['genres']
```

# Analysis genres

```python
genres_list = []
genres_dict = {}
for gen in genres:
    tmp_list = gen.split('|')
    genres_list.append(tmp_list)
    for tmp in tmp_list:
        if tmp in genres_dict:
            genres_dict[tmp] += 1
        else:
            genres_dict[tmp] = 1
```

genres_list

### the number of each genres

```python
genres_dict
```

```python
import matplotlib.pyplot as plt

name = list(genres_dict.keys())
val = list(genres_dict.values())
```

```python
val
```

### Visual

```python
plt.bar(range(len(val)), val, color='r', tick_label=name)
plt.xticks(rotation=90)
plt.show()
```

#### Analysis: Drama and Comedy are the most common.

# Analysis rating

```python
genres_rate_dict = {}

rate_list = {}
for genres in genres_dict:
    genres_rate_dict[genres] = 0.0
    rate_list[genres] = [0,0,0,0,0]
```

```python
for i in range(len(genres_list)):
    print(genres_list[i])
    print(rating[i])
    for tmp in genres_list[i]:
        print(genres_rate_dict[tmp])
#         if rating[i]==np.nan:
#             continue
#         print(rating[i])
#         print(type(float(rating[i])))
        if float(rating[i])>=0.0 and float(rating[i]) <=5.0:
            genres_rate_dict[tmp]+= float(rating[i])
            print(int(float(rating[i])))
            rate_list[tmp][int(float(rating[i]))-1] += 1
```

### For each genre, divide rate into five rates,  show the number of each rate.

```python
rate_list
```

```python
one_list = []
two_list = []
three_list = []
four_list = []
five_list = []
for res in list(rate_list.values()):
    one_list.append(res[0])
    two_list.append(res[1])
    three_list.append(res[2])
    four_list.append(res[3])
    five_list.append(res[4])
```

```python
one_list
```

### Visual 

```python
plt.bar(range(len(one_list)), one_list, label='0-1',fc='r')
plt.bar(range(len(one_list)), two_list,bottom=one_list, label='1-2',fc='g')
plt.bar(range(len(one_list)), three_list,bottom=two_list, label='2-3',fc='b')
plt.bar(range(len(one_list)), four_list, bottom=three_list,label='3-4',fc='y')
plt.bar(range(len(one_list)), five_list, bottom=four_list,label='4-5',fc='black', tick_label=name)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

#### Action and Comedy have higher rating.
